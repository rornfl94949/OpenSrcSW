```python
#실습#1 : 생일 기온 시각화
import matplotlib.pyplot as plt
import csv

f = open('seoul2.csv') ##이전것과 구분하기 위해 2를 붙였습니다.
data = csv.reader(f)
next(data)
max = [] ##최고 기온
min = [] ##최저 기온

for row in data :
    if row[-1] != '' and row[-2] !='':
        if row[0].split('-')[1] == '03' and row[0].split('-')[2] == '23' :
            max.append(float(row[-1]))
            min.append(float(row[-2]))

print(max)
print(min)
plt.plot(max,color='red')
plt.plot(min,color='blue')
plt.show()
```

```python
#실습 #2 최고 기온 시각화

import matplotlib.pyplot as plt
import csv

f = open('seoul2.csv') 
data = csv.reader(f)
next(data)
Jan = [] ## 1월 최고기온
Aug = [] ## 8월 최고기온

for row in data:
    if row[0].split('-')[1] == '01' and row[-1] != '':
        Jan.append(float(row[-1]))
    if row[0].split('-')[1] == '08' and row[-1] != '':
        Aug.append(float(row[-1]))

plt.hist(Aug,bins=100,color='r',label='Aug')
plt.hist(Jan,bins=100,color='b',label='Jan')
plt.legend()
plt.show()
```
