```python
#실습 1번 : 서울의 일교차
import csv
f = open('seoul.csv')
data = csv.reader(f)
header = next(data)
mx = [0,999,-999] #각 값 초기화
mx_date = [''] * 3
label = ['일교차','최저기온','최고기온']
for line in data :
    
    if(line[3] == ''):
        line[3] = 999
    if(line[4] == ''):
        line[4] = -999
    line[3] = float(line[3])
    line[4] = float(line[4])
        
    if line[4] - line[3] > mx[0]:
        mx[0] = line[4] - line[3]
        mx_date[0] = line[0]
        
    if line[3] < mx[1]:
        mx[1] = line[3]
        mx_date[1] = line[0]
        
    if line[4] > mx[2]:
        mx[2] = line[4]
        mx_date[2] = line[0]
     
print("일교차가 가장 큰 날")
str = "날짜 " + mx_date[0] +" 값 : {:.1f}".format(mx[0])
print(str)
print("최저기온이 가장 낮은 날")
str = "날짜 " + mx_date[1] +" 값 : {:.1f}".format(mx[1])
print(str)
print("최고기온이 가장 높은 날")
str = "날짜 " + mx_date[2] +" 값 : {:.1f}".format(mx[2])
print(str)
```

```python
import csv
f = open('subway.csv',encoding='UTF8')
data = csv.reader(f)
header = next(data)
max_use = [0] * 3 # 상위 1등 2등 3등 저장.
min_use = [999999999] * 3 # 하위 1등 2등 3등 저장.
max_station = [''] * 3 # 상위 역 저장
min_station = [''] * 3 # 하위 역 저장
route = '1호선' # 노선 이름 저장. 맨 처음 값이 1호선이므로 1호선 넣음
for line in data :
    if route != line[1] : # 이전 노선이랑 달라지면 그 앞을 출력해줌
        print('노선명 : ' + route)
        print('이용객이 가장 많은 역 순위')
        for i in range(3) :
            str = '{:d}.'.format(i+1)+max_station[i]+'({:d}명)'.format(max_use[i])
            print(str)
        print('이용객이 가장 적은 역 순위')
        for i in range(3) :
            str = '{:d}.'.format(i+1)+min_station[i]+'({:d}명)'.format(min_use[i])
            print(str)
        # 초기화
        print()
        route = line[1]
        for i in range(3) :
            max_use[i] = 0
            min_use[i] = 99999999
            max_station[i] = ''
            min_station[i] = ''
    line[4] = line[4].replace(',','')
    line[5] = line[5].replace(',','')
    use = int(line[4]) + int(line[5])
    
    if use >= max_use[0] : # 1등 역일때
        max_use[2] = max_use[1]
        max_station[2] = max_station[1]
        max_use[1] = max_use[0]
        max_station[1] = max_station[0]
        max_use[0] = use
        max_station[0] = line[3]
    elif use >= max_use[1] : #2등 역일때
        max_use[2] = max_use[1]
        max_station[2] = max_station[1]
        max_use[1] = use
        max_station[1] = line[3]
    elif use >= max_use[2] : #3등 역일때
        max_use[2] = use
        max_station[2] = line[3]
        
    if use <= min_use[0] : # 1등 역일때
        min_use[2] = min_use[1]
        min_station[2] = min_station[1]
        min_use[1] = min_use[0]
        min_station[1] = min_station[0]
        min_use[0] = use
        min_station[0] = line[3]
    elif use <= min_use[1] : #2등 역일때
        min_use[2] = min_use[1]
        min_station[2] = min_station[1]
        min_use[1] = use
        min_station[1] = line[3]
    elif use <= min_use[2] : #3등 역일때
        min_use[2] = use
        min_station[2] = line[3]
        
# 마지막 노선 출력
print('노선명 : ' + route)
print('이용객이 가장 많은 역 순위')
for i in range(3) :
    str = '{:d}.'.format(i+1)+max_station[i]+'({:d}명)'.format(max_use[i])
    print(str)
print('이용객이 가장 적은 역 순위')
for i in range(3) :
    str = '{:d}.'.format(i+1)+min_station[i]+'({:d}명)'.format(min_use[i])
    print(str)   
```
